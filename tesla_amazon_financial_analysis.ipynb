```python
url='https://companiesmarketcap.com/tesla/revenue/'
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = requests.get(url).text

soup = BeautifulSoup(html, "html.parser")
```

```python
table=soup.find_all('table')
```

```python
table = soup.find("table")
```

```python
data = []
```

```python
for row in table.find_all('tr'):
    cells = row.find_all("td")

    if len(cells) >= 2:

            year = cells[0].text.strip()
            revenuee = cells[1].text.strip()
            data.append([year, revenuee])



```

```python
revenue = pd.DataFrame(data, columns=["Year", "Revenue"])
```

```python
print(revenue)
```

```python
revenue["Revenue"] = revenue["Revenue"].str.replace(",", "")
revenue["Revenue"] = revenue["Revenue"].str.replace("$", "", regex=False)
revenue["Year"] = revenue["Year"].str.replace(r"\s*\(.*?\)", "", regex=True)

revenue = revenue[revenue["Revenue"] != ""]
```

```python
print(revenue.head())
```

```python
url2='https://companiesmarketcap.com/nzd/amazon/revenue/'
```

```python
html=requests.get(url2).text
```

```python
soup=BeautifulSoup(html, "html.parser")
```

```python
table=soup.find('table')
```

```python
data2=[]
```

```python
for row in table.find_all('tr'):
    cells = row.find_all("td")
    if len(cells) >= 2:
        year = cells[0].text.strip()
        revenuee = cells[1].text.strip()
        data2.append([year, revenuee])
```

```python
revenue2=pd.DataFrame(data2, columns=["Year", "Revenue"])
```

```python
print(revenue2.head())
```

```python
revenue2["Year"] = revenue2["Year"].str.replace(r"\s*\(.*?\)", "", regex=True)
revenue2["Revenue"] = revenue2["Revenue"].str.replace("NZ$", "", regex=False)
revenue2["Revenue"] = revenue2["Revenue"].str.strip()

```

```python

print(revenue2.head())
```

```python
import yfinance as yf
```

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
```

```python
print(tesla_data.head())
```

```python
amazon = yf.Ticker("AMZN")
amazon_data = amazon.history(period="max")
amazon_data.reset_index(inplace=True)
print(amazon_data.head())
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def make_graph(stock_data, revenue_data, company_name):

    fig = make_subplots(
        rows=2,
        cols=1,
        subplot_titles=(f"{company_name} Stock Price",
                         f"{company_name} Revenue")
    )


    fig.add_trace(
    go.Scatter(
    x=revenue_data["Year"],
    y=revenue_data["Revenue"],
    mode="lines+markers",
    name="Revenue"
),
        row=1,
        col=1
    )

    # نمودار درآمد
    fig.add_trace(
        go.Scatter(
            x=revenue_data["Year"],
            y=revenue_data["Revenue"],
            name="Revenue"
        ),
        row=2,
        col=1
    )

    fig.update_layout(
        height=700,
        showlegend=False,
        title=company_name
    )
    import plotly.io as pio

    pio.renderers.default = "browser"
    fig.show()
```

```python
def convert_revenue(value):


    if isinstance(value, (int, float)):
        return value

    value = value.replace(",", "").strip()

    if value.endswith("T"):
        return float(value[:-1]) * 1_000_000_000_000

    elif value.endswith("B"):
        return float(value[:-1]) * 1_000_000_000

    elif value.endswith("M"):
        return float(value[:-1]) * 1_000_000

    return float(value)
```

```python
revenue["Revenue"] = revenue["Revenue"].apply(convert_revenue)

revenue2["Revenue"] = revenue2["Revenue"].apply(convert_revenue)
```

```python
revenue["Year"] = revenue["Year"].astype(int)
revenue2["Year"] = revenue2["Year"].astype(int)

revenue = revenue.sort_values("Year")
revenue2 = revenue2.sort_values("Year")
```

```python
make_graph(tesla_data, revenue, "Tesla")
```

```python
make_graph(amazon_data, revenue2, "Amazon")
```

```python
revenue.to_csv("data/tesla_revenue.csv", index=False)

revenue2.to_csv("data/amazon_revenue.csv", index=False)

tesla_data.to_csv("data/tesla_stock.csv", index=False)

amazon_data.to_csv("data/amazon_stock.csv", index=False)
```
